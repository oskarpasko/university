# Tablice 

## tablice koincydencji

```r
id <- 1:50
subject <- sample(c("Ang", "Niem", "Jap", "Hisz"), 50, T)
gender <- sample(c("M", "F"), 50, T)
test <- sample(c("Read", "Write", "Listen"), 50, T)
result <- rnorm(50, 60, 15)

obs <- data.frame(id, subject, gender, test, result)

obs
```

```r
mean(obs$result)
```

```r
median(obs$result)
```

```r
quantile(obs$result)
```

```r
quantile(obs$result, probs = seq(0, 1, 0.1))
```

```r
aggregate(obs$result, by = list(gender), FUN = mean) # średni wynik wg płci
```

```r
aggregate(obs$result, by = list(subject, test), FUN=max) # najlepszy wynik wg rodzaju języka i testu
```

```r
aggregate(obs$result, by = list(subject, test), FUN=min) # najgorszy wynik wg rodzaju języka i testu
```

```r
# tabela koincydencji
table(obs$subject, obs$test)
```
