## Szereg rozdzielczy


```r
n <- length(ankieta$Wzrost);
szereg_przedzialowy <- cut(ankieta$Wzrost, sqrt(n))
szereg_przedzialowy
```

```r
table(szereg_przedzialowy)
```

```r
szereg2 <- cut(ankieta$Wzrost, c(155, 180, 200));
table(szereg2)
```

```r
levels(szereg2) <- c("Niski", "Wysoki")
table(szereg2)
```

## Tabele krzyżowe (tabele kontyngencji)

```r
xtabs(~ankieta$`Wykształcenie` + ~ankieta$`Płeć`)
```

```r
table(ankieta$`Wykształcenie`, ankieta$`Płeć`)
```

## Tabele częstości


```r
table(szereg_przedzialowy)
prop.table(table(szereg_przedzialowy))
```

# Wykresy

## Wykres słupkowy

```r
barplot(prop.table(table(szereg_przedzialowy)), 
        main="Procent badanych", 
        xlab="Przedział wzrostu", 
        ylab="Odsetek ankietowanych", 
        axes=TRUE,
        border="red", density=c(10,15))
```

## Skumulowane wykresy słupkowe

```r
set.seed(102)
data <- matrix(sample(1:30, 15), nrow=3)
colnames(data) <- c("A", "B", "C", "D", "E")
rownames(data) <- c("X", "Y", "Z")
data
```

```r
barplot(data,
        col=colors()[c(23, 89, 12)],
        border="white",
        font.axis=2,
        beside=T,
        legend=rownames(data),
        font.lab=2)
```
